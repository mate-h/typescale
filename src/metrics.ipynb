```python
# metrics
# UPM: 2048
# Variable axis opsz [14–32]
# Variable axis wght [100–900]
# Constant axis ital [0, 9.4°]
# Style classification: Sans-serif, Grotesk
# x-height, opsz=14: 1118 UPM
# x-height, opsz=32: 1056 UPM
# Cap height: 1490 UPM
# Ascender: 1984 UPM
# Descender: -494 UPM
# Vendor ID: RSMS

# type scale
# theme: {
#     fontSize: {
#       h1: '2.986rem',
#       h2: '2.488rem',
#       h3: '2.074rem',
#       h4: '1.728rem',
#       h5: '1.44rem',
#       h6: '1.2rem',
#       base: '1rem',
#       sm: '0.833rem',
#       xs: '0.694rem',
#     },
#   },
```

```python
import json
import math

# Font metrics
upm = 2048
x_height_upm = 1118
default_font_size = 16  # px
desired_grid = 4  # px

# Calculate x-height in pixels for the default font size
x_height_px = (default_font_size * x_height_upm) / upm

# Find the nearest multiple of the desired grid
nearest_grid_x_height = round(x_height_px / desired_grid) * desired_grid

# Calculate the new base font size to achieve the grid-aligned x-height
new_base_size_px = (default_font_size * nearest_grid_x_height) / x_height_px
new_base_size_rem = new_base_size_px / default_font_size

# Original scale factors
scale_factors = {
    'h1': 2.986,
    'h2': 2.488,
    'h3': 2.074,
    'h4': 1.728,
    'h5': 1.44,
    'h6': 1.2,
    'base': 1,
    'sm': 0.833,
    'xs': 0.694
}

# Calculate new font sizes
font_sizes = {key: round(new_base_size_rem * factor, 3) for key, factor in scale_factors.items()}

# Create theme dictionary
theme = {
    'fontSize': {key: f'{value}rem' for key, value in font_sizes.items()}
}

# Print the result using json module
print(json.dumps({'theme': theme}, indent=2))
```
